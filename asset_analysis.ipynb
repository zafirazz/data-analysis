## Conversion funnels
In the given dataset, a conversion funnels to Assets will be divided for Personal and Business customers.

For Personal Customers:
* PROFILE_CREATED
* FIRST_COMPLETED_ACTION
* FIRST_ACCOUNT_DETAILS
* FIRST_DIRECT_DEBIT
* FIRST_ASSETS

For Business Customers:
* PROFILE_CREATED
* FIRST_COMPLETED_ACTION
* FIRST_ACCOUNT_DETAILS
* FIRST_DIRECT_DEBIT
* FIRST_ASSETs
  

```python
conversion = ['PROFILE_CREATED', 'FIRST_COMPLETED_ACTION', 'FIRST_ACCOUNT_DETAILS', 'FIRST_DIRECT_DEBIT', 'FIRST_ASSETS']

conversion_rates_personal = {}
conversion_rates_business = {}

for step in conversion:
    personal_completed_step = personal_customers[step].notnull().sum()
    personal_prev_step = personal_customers[conversion[conversion.index(step) - 1]].notnull().sum() if conversion.index(step) > 0 else len(personal_customers)
    
    conversion_rate_personal = (personal_completed_step / personal_prev_step) * 100
    conversion_rates_personal[step] = conversion_rate_personal

    business_completed_step = business_customers[step].notnull().sum()
    business_prev_step = business_customers[conversion[conversion.index(step) - 1]].notnull().sum() if conversion.index(step) > 0 else len(business_customers)
    
    conversion_rate_business = (business_completed_step / business_prev_step) * 100
    conversion_rates_business[step] = conversion_rate_business

print("Conversion Rates for Personal Customer: ")
for step, rate in conversion_rates_personal.items():
    print(f"{step}: {rate:.2f}%")

print("\nConversion Rates for Business Customers:")
for step, rate in conversion_rates_business.items():
    print(f"{step}: {rate:.2f}%")
```

```python
personal_customers = personal_customers.drop_duplicates(subset=['PROFILE_CREATED'])
business_customers = business_customers.drop_duplicates(subset=['PROFILE_CREATED'])

funnels = {'PROFILE_CREATED': 'FIRST_COMPLETED_ACTION',
          'FIRST_COMPLETED_ACTION': 'FIRST_ACCOUNT_DETAILS',
          'FIRST_ACCOUNT_DETAILS': 'FIRST_DIRECT_DEBIT',
          'FIRST_DIRECT_DEBIT': 'FIRST_ASSETS'}

for first, second in funnels.items():
    try:
        not_null_personal = pd.notnull(personal_customers[first]) & pd.notnull(personal_customers[second])
        not_null_business = pd.notnull(business_customers[first]) & pd.notnull(business_customers[second])
        if not_null_personal.any():
            personal_customers[first] = pd.to_datetime(personal_customers[first])
            personal_customers[second] = pd.to_datetime(personal_customers[second])

            time_difference_p = personal_customers.loc[not_null_personal, second] - personal_customers.loc[not_null_personal, first]

            avg_time_p = time_difference_p.mean()

            print(f"The average time for Personal customers from {first} to {second}: ", avg_time_p)
        else:
            print(f"Skipping calculation due to null values.")

        if not_null_business.any():
            business_customers[first] = pd.to_datetime(business_customers[first])
            business_customers[second] = pd.to_datetime(business_customers[second])

            time_difference_b = business_customers.loc[not_null_business, second] - business_customers.loc[not_null_business, first]

            avg_time_b = time_difference_b.mean()

            print(f"The average time for Business customers from {first} to {second}: ", avg_time_b)
        else:
            print(f"Skipping calculation due to null values.")
    except Exception as e:
        print(f"Error occurred during calculation: {e}")

    
```

## Average time customers spend at each step

Now I will investigate the average time that customers spend at each step.

The average time for Personal customers from PROFILE_CREATED to FIRST_COMPLETED_ACTION is approximately 9 days and 21 hours, while for Business customers, it is around 7 days and 13 hours. This indicates that Personal customers tend to take slightly longer to complete registration process compared to Business customers. The reasons behind this discrepancy could be attributed to differing documentation requirements, decision-making processes, or user experience optimizations tailored to each segment. Same situation for FIRST_COMPLETED_ACTION to FIRST_ACCOUNT DETAILS: Personal customers take 34 days and 21 hours on average, while Business customers, do it faster with difference of 10 days.

Now the interesting part is the time taken for transition from FIRST_ACCOUNT_DETAILS to FIRST_DIRECT_DEBIT and FIRST_DIRECT_DEBIT to FIRST_ASSETS.In case of first account details, Personal customers are faster comparing to Business customers. 

While for FIRST_DIRECT_DEBIT it seems 0 days and negative number days might be attributed that those customers who did not have first direct debit, had Assets which is an end goal for this particular project. 

Below I investigate this theory.

```python
segments = {
    'All Customers': customer_data,
    'Personal Customers': personal_customers,
    'BUsiness Customers': business_customers
}

for name, data in segments.items():
    customers_with_assets = data[data['FIRST_ASSETS'].notnull()]
    customers_assets_nodebit = customers_with_assets[customers_with_assets['FIRST_DIRECT_DEBIT'].isnull()]
    
    total_customers_with_assets = customers_with_assets.shape[0]
    customers_assets_nodebit_count = customers_assets_nodebit.shape[0]
    percentage = (customers_assets_nodebit_count / total_customers_with_assets) * 100

    print(f"The percentage of {segment_name} who have Assets, but not First Direct Debit: {percentage:.2f}%")
```
